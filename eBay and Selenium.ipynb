## eBay and Selenium

```python
# Import packages
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import time
```

### Automate Browser to eBay

```python
## Navigate to the page and search for results
# Start a google chrome page
s = Service('chromedriver-mac-arm64/chromedriver')
driver = webdriver.Chrome(service=s)

# Pause between requests
time.sleep(5)

# Navigate to ebay
driver.get('https://www.ebay.com/')

# Pause between requests
time.sleep(5)

# Extract the search box and search for "Cell Phones"
input = driver.find_element(By.CLASS_NAME, 'gh-tb.ui-autocomplete-input')
input.send_keys('Cell Phones\n')

# Pause between requests
time.sleep(5)


## Apply filters
# Click on Network: Unlocked
unlocked = driver.find_element(By.XPATH, '//*[@aria-label="Unlocked"]')
unlocked.click()
time.sleep(5)

# Click on Brand: LG
LG = driver.find_element(By.XPATH, '//*[@aria-label="LG"]')
LG.click()
time.sleep(5)

# Click on Screen Size: 6 in or More
size = driver.find_element(By.XPATH, '//*[@aria-label="6 in or More"]')
size.click()
time.sleep(5)

# Click on Storage capacity: 128 GB
storage = driver.find_element(By.XPATH, '//*[@aria-label="128 GB"]')
storage.click()
time.sleep(5)

# Click on Lock Status: Factory Unlocked
lock = driver.find_element(By.XPATH, '//*[@aria-label="Factory Unlocked"]')
lock.click()
time.sleep(5)

# Click on Condition: Used
condition = driver.find_element(By.XPATH, '//*[@aria-label="Used"]')
condition.click()
time.sleep(5)


## Save the resulting page
# Save it to a html file
html_content = driver.page_source
with open("unlocked-lg-128gb-used.html", "w", encoding="utf-8") as file:
    file.write(html_content)
    
# Quit driver
time.sleep(10)
driver.quit()
```

### Analyze Data and Print Results

```python
# Set working directory
directory = '../Individual Project 2'

# Read the html file
with open("unlocked-lg-128gb-used.html", "r", encoding="utf-8") as file:
    html = file.read()

# Use Beautifulsoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Print the page
# print(soup.prettify)
```

```python
# Find all product information
products = soup.select('div.s-item__info')
len(products)
```

```python
# Extract and print infromation from the product information
for product in products[1:]:
    
    # Titles
    titles = product.select('div.s-item__title > span')
    for title in titles:
        print("Title: ", title.text)
    
    # Sellers
    sellers = product.select('span.s-item__seller-info')
    for seller in sellers:
        print("Seller and seller rating: ", seller.text)
    
    # Prices
    prices = product.select('span.s-item__price')
    for price in prices:
        print("Price: ", price.text)
    
    # Shipping
    ships = product.select('span.s-item__shipping')
    for ship in ships:
        print("Shipping information: ", ship.text)
    
    # Return
    returns = product.select('span.s-item__free-returns')
    for re in returns:
        print("Return information: ", re.text)
    
    # Bidding
    bids = product.select('span.s-item__bids')
    for bid in bids:
        print("Number of bids: ", bid.text)
    
    print("----------------------------------------------------------------------------------------------")
```
